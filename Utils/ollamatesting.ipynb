```python
import json
import requests
from OllamaAPI import *
```

```python
def listsModels(url = "http://localhost:11434/api/tags"):
    
    response = requests.get(url)

    modelList = []

    if response:

        try:
            j = response.json()

            #If models 
            if "models" in j:
                for m in j["models"]:
                    modelList.append(m["name"])
            else:
                print(f"Failed to retrieve models from {response}")
                return []
            
        except json.JSONDecodeError:
            print(f"Failed to retrieve models from {response}")
            return []
```
